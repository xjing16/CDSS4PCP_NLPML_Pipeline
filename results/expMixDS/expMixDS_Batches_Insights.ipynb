```python
import pandas as pd
import matplotlib as plt
```

```python
# df = pd.read_csv('expMixDS.csv', sep=';')
df = pd.concat((pd.read_csv(file, sep=';') for file in ['expMixDS_Batch1.csv', 'expMixDS_Batch2.csv','expMixDS_Batch3.csv','expMixDS_Batch4.csv','expMixDS_Batch5.csv']))
```

```python
df.query("`Test DataSet` == 'GS91'").groupby('Combination')[['Precision','Recall','F1Score','Accuracy']].describe().transpose().round(2)
```

```python
df.query("`Test DataSet` == 'GS91'").groupby('Combination')[['Precision','Recall','F1Score','Accuracy']].median().transpose().round(2)
```

```python
df.query("`Test DataSet` == 'GS91'").groupby('Combination')[['Precision','Recall','F1Score','Accuracy']].var().transpose().round(2)
```

```python
plt.rcParams["figure.figsize"] = (20,6)
df.query("`Test DataSet` == 'GS91'").boxplot(column='Precision',by='Combination',showfliers=False)
```
